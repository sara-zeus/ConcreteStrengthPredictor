```python
import tensorflow as tf
from tensorflow import keras

from keras.models import Sequential  # Note the capital 'S' in 'Sequential'
from keras.layers import Dense


import pandas as pd
```

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

checking how many data points we have

```python
concrete_data.shape
```

```python
concrete_data.describe()
```

checking for null values

```python
concrete_data.isnull().sum()
```

data is pretty clean

The target variable in this problem is the concrete sample strength. Therefore, our predictors will be all the other columns.


```python
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']]
target = concrete_data["Strength"]
```

Checking to make sure we have the predictors and target ready for the model

```python
predictors.head()
```

```python
target.head()
```

now I normalize the data by substracting the mean and dividing by the standard deviation.


```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

Number of predictors


```python
n_cols = predictors_norm.shape[1]
n_cols
```

****Building a Neural Network****

function below creates two hidden layers, each of 50 hidden units.

```python
def regression_model():

    model = Sequential()
    model.add(Dense(50, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(1))


    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

now I train and test the network

```python
model = regression_model()
```


Subsequently, we'll simultaneously train and test the model by employing the fit method. To ensure a 30% validation set, we'll reserve a portion of the data. The model will undergo training over 100 epochs.

```python
model.fit(predictors_norm, target, validation_split=0.3, epochs=100, verbose=2)
```
